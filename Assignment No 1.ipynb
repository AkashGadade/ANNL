```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def liner(x):
    return x

inputs = np.array([x for x in range(-10,10)])
output  = np.array([liner(x) for x in inputs])
plt.title("linear activation")

plt.plot(inputs,output)
```

```python
def stepfun(x):
    if x>0:
        return 1
    else:
        return 0

inputs = np.array([x for x in range(-50,50)])
output  = np.array([stepfun(x) for x in inputs])
plt.title("linear activation")

plt.plot(inputs,output)
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
inputs = np.array([x for x in range(-10,10)])
output  = np.array([sigmoid(x) for x in inputs])
plt.title("linear activation")

plt.plot(inputs,output)

```

```python
def relu(x):
    return max(x,0)
inputs = np.array([x for x in range(-10,10)])
output  = np.array([relu(x) for x in inputs])
plt.title("linear activation")

plt.plot(inputs,output)
```

```python
def tanh(x):
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
inputs = np.array([x for x in range(-10,10)])
output  = np.array([tanh(x) for x in inputs])
plt.title("linear activation")

plt.plot(inputs,output)
```

```python
def leakRelu(x,alpha=0.1):
    return np.maximum(alpha*x,x)

inputs = np.array([x for x in range(-10,10)])
output  = np.array([leakRelu(x) for x in inputs])
plt.title("linear activation")

plt.plot(inputs,output)
```
